**PROJECT TITLE:** Marketing Campaign for Banking Products

**DATA DESCRIPTION:** The data file contains data on 5000 customers. The data include customer
demographic information (age, income, etc.), the customer's relationship with the bank
(mortgage, securities account, etc.), and the customer response to the last personal
loan campaign (Personal Loan).
Among these 5000 customers, only 480 (= 9.6%) accepted the personal loan that was
offered to them in the earlier campaign. 

**CONTEXT:** The bank has a growing customer base. The bank wants to increase borrowers (asset
customers) base to bring in more loan business and earn more through the interest on
loans. So , the bank wants to convert the liability based customers to personal loan
customers. (while retaining them as depositors). A campaign that the bank ran last year
for liability customers showed a healthy conversion rate of over 9% success. The
department wants you to build a model that will help them identify the potential
customers who have a higher probability of purchasing the loan. This will increase the
success ratio while at the same time reduce the cost of the campaign.

**ATTRIBUTE INFORMATION:** 
*   ID: Customer ID
*   Age: Customer's age in years
*   Experience: Years of professional experience
*   Income: Annual income of the customer
*   ZIP Code: Home Address ZIP code
*   Family: Family size of the customer
*   CCAvg: Average spending on credit cards per month
*   Education: Education Level: 1: Undergrad; 2: Graduate; 3: Advanced/Professional
*   Mortgage: Value of house mortgage if any
*   Personal Loan: Did this customer accept personal loan offered in the last campaign?
*   Securities Account: Does the customer have a securities account with the bank?
*   CD Account: Does the customer have a certificate of deposit (CD) account with the bank?
*   Online: Does the customer use internet banking facilities?
*   Credit card: Does the customer use a credit card issued by the bank?

**OBJECTIVE:** The classification goal is to predict the likelihood of a liability customer buying personal
loans by building a logistic regression model and also comparing this with other models.

**STEP 2: CHECKING IF THE DATASET HAS TO BE CLEANED**

The Experience column has a few negative entries. As one's experience cannot be a negative value, they must be replaced. Here we will replace it with the median experience value.

```python
# Replacing the negative values in the Experience column with the median value.

for x in data['Experience']:
  if x<0:
    data['Experience'].replace(to_replace=x, value=data['Experience'].median(), inplace=True, limit=None, regex=False, method='pad')

data.describe()
```

```python
# Dropping columns which are not significant from the dataset.

data_modified = data.drop(['ID', 'Experience', 'ZIP Code', 'Securities Account', 'Online', 'CreditCard', 'Family'], axis=1)
data_modified.head(10)
```

**STEP 4: APPLYING NECESSARY TRANSFORMATIONS FOR THE FEATURE VARIABLES**

As seen from earlier graphs plotted using continuous variables, there are columns whose distributions are skewed. Hence a transformation technique has to be applied to Income, CCAvg and Mortgage columns.

Here we are applying quantile transformation scaler as it converts the variable distribution to a normal distribution and scales it accordingly. Since it makes the variable normally distributed, it also deals with outliers.

```python
# Applying quantile transformation.

scaler = QuantileTransformer( )
scaler.fit(data_modified['Income'].values.reshape(-1,1))
data_modified['Income'] = scaler.fit_transform(data_modified['Income'].values.reshape(-1,1))
sns.distplot(data_modified['Income'])
```

```python
scaler = QuantileTransformer()
scaler.fit(data_modified['CCAvg'].values.reshape(-1,1))
data_modified['CCAvg'] = scaler.fit_transform(data_modified['CCAvg'].values.reshape(-1,1))
sns.distplot(data_modified['CCAvg'])
```

```python
scaler = QuantileTransformer()
scaler.fit(data_modified['Mortgage'].values.reshape(-1,1))
data_modified['Mortgage'] = scaler.fit_transform(data_modified['Mortgage'].values.reshape(-1,1))
sns.distplot(data_modified['Mortgage'])
```

```python
# To check for outliers after applying transformation.

data_modified.drop(['Personal Loan','Age','Education','CD Account'],axis=1).plot.box(subplots=True,layout=(1,3),figsize=(15,5))
```

```python
data_modified.head(10)
```

**STEP 5: NORMALISATION AND SPLITTING OF DATA INTO TRAINING AND TEST SETS IN THE RATIO OF 70:30**

In the previous step, when quantile transformation was applied, the dataset was also normalised. Hence there is no need to apply any explicit technique for normalisation once again.

```python
# Dividing the dataset into input parameters and target variable.

X = data_modified.drop(['Personal Loan'],axis=1)
y = data_modified['Personal Loan']
```

As seen earlier, the Personal Loan column is highly unevenly distributed. Therefore the sample has to be split as a stratified sample.

```python
# Splitting the dataset into training and test sets as a stratified sample

X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**STEP 6: BUILDING A LOGISTIC REGRESSION MODEL**

```python
# Fitting the model.

logistic_regression= LogisticRegression(random_state=1)
logistic_regression.fit(X_train,y_train)
```

```python
# Prediction and Performance on Test Data.

from sklearn.metrics import confusion_matrix
y_pred_test=logistic_regression.predict(X_test)
print('Predicted values using test data:',y_pred_test)
print('Accuracy: ',metrics.accuracy_score(y_test, y_pred_test)*100,'%')
cm = confusion_matrix(y_test,y_pred_test)
print('Confusion matrix:')
print(cm)
```

To check if the model has undergone overfitting we shall predict using the train data.

```python
# Prediction and Performance on Train Data.

from sklearn.metrics import confusion_matrix
y_pred_train = logistic_regression.predict(X_train)
print('Predicted values using train data:',y_pred_train)
print('Accuracy: ',metrics.accuracy_score(y_train, y_pred_train)*100,'%')
cm = confusion_matrix(y_train, y_pred_train)
print('Confusion matrix:')
print(cm)
```

Observation: The accuracies of predicted data using train and test datasets are almost equal. Hence we can conclude that our model works fine.

**STEP 7: PRINTING THE METRICS FOR EVALUATING THE MODEL PERFORMANCE**

```python
# Printing performance summary for the model based on test data.

print("Accuracy: ", metrics.accuracy_score(y_test,y_pred_test)*100,"%\n")
cm = confusion_matrix(y_test,y_pred_test)
print('Confusion matrix: \n',cm,'\n')
print("Classification Report for Logistic Regression model: \n",metrics.classification_report(y_test, y_pred_test))
```

```python
# Plotting a confusion matrix.

skplt.metrics.plot_confusion_matrix(y_test, y_pred_test, cmap='Reds')
```

```python
# Printing performance summary for the model based on train data.

print("Accuracy: ", metrics.accuracy_score(y_pred_train,y_train)*100,"%\n")
cm = confusion_matrix(y_train,y_pred_train)
print('Confusion matrix: \n',cm,'\n')
print("Classification Report for Logistic Regression model: \n",metrics.classification_report(y_train, y_pred_train))
```

Observation: The model is highly accurate with an accuracy of 95.27% for test data and 95.51% for train data and also with high recall value and f1 scores.
